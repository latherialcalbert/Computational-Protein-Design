```python
import pandas as pd
import glob
from scipy.optimize import minimize_scalar

def filter_dataframe(df, column, criteria, direction):
    """
    Filter the DataFrame based on the specified column, criteria, and direction.

    Parameters:
    - df: pandas DataFrame
    - column: str, column name
    - criteria: numeric value, filtering criteria
    - direction: str, 'above', 'below', or 'equal', direction of filtering

    Returns:
    - filtered_df: pandas DataFrame, filtered copy of the original DataFrame
    """
    filtered_df = df.copy()
    if direction == 'above':
        filtered_df = filtered_df[filtered_df[column] > criteria]
    elif direction == 'below':
        filtered_df = filtered_df[filtered_df[column] < criteria]
    elif direction == 'equal':
        filtered_df = filtered_df[filtered_df[column] == criteria]
    return filtered_df

# Read CSV file in the base directory
csv_files = glob.glob('*.csv')
if len(csv_files) == 1:
    csv_filename = csv_files[0]
    df = pd.read_csv(csv_filename)
else:
    print("Error: There should be exactly one CSV file in the base directory.")
    exit()

# Specify column headers for filtering
columns = ['pae_interaction_scores', 'dSASA_int', 'dG_cross', 'sc_value', 'delta_unsatHbonds']

# Define the desired target range
target_range = (10, 50)  # Adjust as needed

# Specify the desired filter directions for each column
filter_directions = ['below', 'above', 'below', 'above', 'above']  # Adjust as needed

# Initialize filter_values
filter_values = []

# Optimize filter_values based on the desired target range
for i, column in enumerate(columns):
    direction = filter_directions[i]
    column_values = df[column]

    if direction == 'below':
        # Find the global minimum value within the target range
        result = minimize_scalar(lambda x: len(filter_dataframe(df, column, x, direction)), bounds=(column_values.min(), column_values.max()), method='bounded')
        criteria = result.x

    elif direction == 'above':
        # Find the global maximum value within the target range
        result = minimize_scalar(lambda x: -len(filter_dataframe(df, column, x, direction)), bounds=(column_values.min(), column_values.max()), method='bounded')
        criteria = result.x

    filter_values.append(criteria)

    # Perform filtering using the optimized filter_values and filter_directions
    df = filter_dataframe(df, column, criteria, direction)

# Check if the filtered DataFrame is within the target range
num_rows = len(df)
if not (target_range[0] <= num_rows <= target_range[1]):
    print("Warning: The filtered DataFrame is not within the target range.")

# Display the final filtered DataFrame
print(df)
```

```python
import pandas as pd
import glob
from scipy.optimize import minimize

def filter_dataframe(df, column, criteria, direction):
    """
    Filter the DataFrame based on the specified column, criteria, and direction.

    Parameters:
    - df: pandas DataFrame
    - column: str, column name
    - criteria: numeric value, filtering criteria
    - direction: str, 'above', 'below', or 'equal', direction of filtering

    Returns:
    - filtered_df: pandas DataFrame, filtered copy of the original DataFrame
    """
    filtered_df = df.copy()
    if direction == 'above':
        filtered_df = filtered_df[filtered_df[column] > criteria]
    elif direction == 'below':
        filtered_df = filtered_df[filtered_df[column] < criteria]
    elif direction == 'equal':
        filtered_df = filtered_df[filtered_df[column] == criteria]
    return filtered_df

# Read CSV file in the base directory
csv_files = glob.glob('*.csv')
if len(csv_files) == 1:
    csv_filename = csv_files[0]
    df = pd.read_csv(csv_filename)
else:
    print("Error: There should be exactly one CSV file in the base directory.")
    exit()

# Specify column headers for filtering
columns = ['pae_interaction_scores', 'dSASA_int', 'dG_cross/dSASAx100', 'sc_value', 'delta_unsatHbonds']

# Define the desired target range
target_range = (10, 100)  # Adjust as needed

# Specify the desired filter directions for each column
filter_directions = ['below', 'above', 'below', 'above', 'below']  # Adjust as needed

# Initialize filter_values
filter_values = []

# Optimize filter_values based on the desired target range using gradient descent
for i, column in enumerate(columns):
    direction = filter_directions[i]
    column_values = df[column].valuess

    if direction == 'below':
        # Define the objective function for minimization
        def objective(x):
            filtered_df = filter_dataframe(df, column, x[0], direction)
            return abs(target_range[0] - len(filtered_df))

    elif direction == 'above':
        # Define the objective function for maximization
        def objective(x):
            filtered_df = filter_dataframe(df, column, x[0], direction)
            return abs(target_range[1] - len(filtered_df))

    # Set the initial guess for the filter value
    x0 = [column_values.mean()]

    # Perform gradient descent optimization
    result = minimize(objective, x0, method='L-BFGS-B', bounds=[(column_values.min(), column_values.max())])
    criteria = result.x[0]

    filter_values.append(criteria)

    # Perform filtering using the optimized filter_values and filter_directions
    df = filter_dataframe(df, column, criteria, direction)

# Check if the filtered DataFrame is within the target range
num_rows = len(df)
if not (target_range[0] <= num_rows <= target_range[1]):
    print("Warning: The filtered DataFrame is not within the target range.")

# # Print the filtered values
# print("Filtered Values:")
# print(df)

# Print the filtered columns
print("filter values are:")
print(filter_values)
print(filter_directions)
filtered_columns = df[columns]
print("Filtered Columns:")
print(filtered_columns)
num_rows = len(df)
print("Number of Rows:", num_rows)

```
